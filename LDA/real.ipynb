```python
new_words = {''}
words = new_words.union(thai_words())
custom_dictionary_trie = dict_trie(words)

vowels = ('”','“','”','_','!!','–','-','(',')','!','"')
stopword_set = frozenset(vowels)
TH_stopword = thai_stopwords().union(stopword_set)

p_stemmer = PorterStemmer()
def clean(word):
    dfTitle = word.strip('!()/\|}"’‘{_<>[]')
    tokendfTitle = word_tokenize(normalize(dfTitle), custom_dict=custom_dictionary_trie, keep_whitespace=False, engine="newmm")
    Word_in_Title = [word for word in tokendfTitle if not word in TH_stopword]
    Word_in_Title = [p_stemmer.stem(i)  for i in Word_in_Title]
    return  ''.join(Word_in_Title)
```

```python
def text_tokenizer(text):
  terms = [k.strip() for k in word_tokenize(text) if len(k.strip()) > 0 and k.strip() not in stop_words]
  return [t for t in terms if len(t) > 0 or t is not None]

def text_processor(text: str) -> str:
  omit_key = [ord(c) for c in '*?!&']
  omit_value = [None] * len(omit_key)
  omit = dict(zip(omit_key, omit_value))
 
  return ' '.join([t.strip() \
    for t in text.translate(omit).split("\n") \
  ])

stop_words = thai_stopwords()
```

```python
df.text = df.text.astype(str)
df.text = df.text.apply(lambda x: x.encode('utf-8').strip())
```

```python
stop_words = [t.encode('utf-8') for t in list(thai_stopwords())]

tfidf_vectors = TfidfVectorizer(
    tokenizer=text_tokenizer,
    preprocessor=text_processor,
    ngram_range=(2, 3),
    
    stop_words=stop_words,
    min_df=int(df.shape[0] * 0.01),
    max_features=3_000
)
```

```python
X = tfidf_vectors.fit_transform(df.text)
```

```python
terms_tfidf = pd.DataFrame(
    X.toarray(),
    columns=tfidf_vectors.get_feature_names(),
    index=df.index,

)

terms_tfidf.sample(n=5)
```

```python
terms_tfidf.sum().sort_values(ascending=False).head(20)
```

```python
lda_topics_modeling = LDA(
    n_components=5,
    learning_decay=0.5,
    n_jobs=-1,
    random_state=1234
)

topics = lda_topics_modeling.fit_transform(X)
```

```python
topics_df = pd.DataFrame(
    lda_topics_modeling.components_,
    columns=terms_tfidf.columns,
    index=[f'Topic {i+1}' for i in range(topics.shape[1])]
).T
```

```python
for t in topics_df.columns:
  terms = [k.replace(' ', '') for k in topics_df.sort_values(by=t, ascending=False)[[t]].head(10).index.to_list()]

  print(f'{t}:', "\n", ', '.join(terms))
```
